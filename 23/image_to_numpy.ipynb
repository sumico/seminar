```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the image
image = Image.open("/content/drive/MyDrive/1.jpg")
```

```python
# Convert to NumPy array
image_array = np.array(image)

# Print array shape and data type
print("Array shape:", image_array.shape)
print("Array data type:", image_array.dtype)
```

```python
# Visualize the image
plt.imshow(image_array)
plt.axis('off')
plt.title("Original Image")
plt.show()
```

```python
# Plot histogram for each RGB channel
colors = ['red', 'green', 'blue']
for i, color in enumerate(colors):
    plt.hist(image_array[:, :, i].ravel(), bins=256, color=color, alpha=0.5, label=f'{color} channel')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.title('Histogram of RGB Channels')
plt.legend()
plt.show()
```
