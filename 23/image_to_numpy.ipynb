```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

###ステップ1：画像のアップロード/Step 1: Upload images

```python
# The file already exists in Google Drive, so we don't need to upload it.
# We can directly access it using its path.
image_path = '/content/drive/MyDrive/1.jpg'
```

###ステップ2：画像をNumPy配列に変換/Step 2: Convert the image to a NumPy array

```python
# 画像を開く/Open image
image = Image.open(image_path)

# NumPy配列に変換/Convert to NumPy array
image_array = np.array(image)

# 配列の形状を表示/Display array shape
print("NumPy array shape of the image:", image_array.shape)
```

###ステップ3：画像を表示して確認/Step 3: Display and confirm the image

NumPy配列に変換された画像を表示します。/Displays images converted to NumPy arrays.


```python
plt.imshow(image_array)
plt.axis('off')  # 軸を非表示にする/Hide axis
plt.title("Uploaded image")
plt.show()
```
